```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
print("Hello World")
```

```python
df_food=pd.read_csv(r'F:\train.csv\train.csv')
```

```python
df_food.head()
```

```python
df_center=pd.read_csv(r'F:\fulfilment_center.csv\fulfilment_center_info.csv')
```

```python
df_center.head()
```

```python
df_meal=pd.read_csv(r'F:\train_GzS76OK\meal_info.csv')
```

```python
df_meal.head()
```

```python
df=pd.merge(df_center, df_food, how='outer', on='center_id')
```

```python
df.head()
```

```python
df=pd.merge(df, df_meal,how='outer', on='meal_id')
```

```python
df.head()
```

```python
table=pd.pivot_table(data=df,index='category',values='num_orders', aggfunc=np.sum)
```

```python
plt.bar(table.index,table['num_orders'])
plt.xticks(rotation=70)
plt.title("Popularity of foods")
plt.xlabel('Food category')
plt.ylabel('Number of orders')
plt.show()
```

```python
df['revenue']=df.apply(lambda x:x.checkout_price*x.num_orders,axis=1)
df['month']=df['week'].apply(lambda x:x//4)
```

```python
df.head()
```

```python
month=[]
month_revenue=[]
for i in range(max(df['month'])):
    month.append(i)
    month_revenue.append(df[df['month']==i].revenue.sum())
week=[]
week_revenue=[]
for i in range(max(df['week'])):
    week.append(i)
    week_revenue.append(df[df['week']==i].revenue.sum())
```

```python
fig, ax=plt.subplots(nrows=1, ncols=2, figsize=(20,5))
fig.suptitle("Revenue Observation")
ax[0].plot(month,month_revenue)
ax[0].set_title("Monthly revenue")
ax[0].set_xlabel('Month')
ax[0].set_ylabel('Revenue')

ax[1].plot(week, week_revenue)
ax[1].set_title("Weekly revenue")
ax[1].set_xlabel("Week")
ax[1].set_ylabel("Revenue")
```
